```python
# Estruturas de dados para armazenar resultados intermediários
import numpy as np

# Inicializando uma matriz 2D para armazenar o progresso dos estudantes em diferentes treinamentos
# Supondo que temos 10 alunos e 5 treinamentos (colunas: progresso de 0 a 100% em cada treinamento)
progresso_treinamento = np.zeros((10, 5))  # 10 alunos, 5 treinamentos

# Usando um dicionário para armazenar detalhes sobre cada aluno e o tipo de treinamento realizado
treinamentos_detalhados = {
    "aluno_1": {"treinamento": "cirurgia laparoscópica", "resultados": []},
    "aluno_2": {"treinamento": "procedimento endoscópico", "resultados": []},
    # ... outros alunos
}

# Funções que implementam a lógica da solução

def inicializar_treinamento(aluno_id, treinamento_tipo):
    """
    Inicializa um novo treinamento para um aluno específico.
    Atualiza o dicionário de treinamentos detalhados com o tipo de treinamento.
    """
    treinamentos_detalhados[f"aluno_{aluno_id}"] = {
        "treinamento": treinamento_tipo,
        "resultados": []
    }
    print(f"Treinamento '{treinamento_tipo}' inicializado para o aluno {aluno_id}.")

def atualizar_progresso(aluno_id, treinamento_id, progresso):
    """
    Atualiza o progresso do aluno em um treinamento específico.
    Atualiza tanto a matriz de progresso quanto o dicionário com resultados detalhados.
    """
    progresso_treinamento[aluno_id-1, treinamento_id-1] = progresso
    treinamentos_detalhados[f"aluno_{aluno_id}"]["resultados"].append(
        {"treinamento_id": treinamento_id, "progresso": progresso}
    )
    print(f"Progresso do aluno {aluno_id} no treinamento {treinamento_id} atualizado para {progresso}%.")

def finalizar_treinamento(aluno_id, treinamento_id):
    """
    Marca o treinamento de um aluno como finalizado.
    Adiciona uma marcação de conclusão no dicionário detalhado.
    """
    treinamentos_detalhados[f"aluno_{aluno_id}"]["resultados"][-1]["status"] = "finalizado"
    print(f"Treinamento {treinamento_id} do aluno {aluno_id} finalizado.")

# Função para exibir os resultados intermediários
def exibir_resultados():
    """
    Exibe o progresso e status detalhado dos treinamentos de todos os alunos.
    """
    print("\n--- Resultados Detalhados dos Treinamentos ---")
    for aluno, detalhes in treinamentos_detalhados.items():
        print(f"\n{aluno}: Treinamento - {detalhes['treinamento']}")
        for resultado in detalhes["resultados"]:
            progresso = resultado.get("progresso", 0)
            status = resultado.get("status", "em andamento")
            print(f"   Treinamento ID: {resultado['treinamento_id']}, Progresso: {progresso}%, Status: {status}")

# Simulação de uso do sistema para 3 alunos e 2 treinamentos
# Aluno 1 participa de um treinamento de cirurgia laparoscópica e endoscopia
inicializar_treinamento(1, "cirurgia laparoscópica")
atualizar_progresso(1, 1, 50)  # Progresso de 50%
finalizar_treinamento(1, 1)    # Finalizando treinamento 1
inicializar_treinamento(1, "endoscopia")
atualizar_progresso(1, 2, 70)  # Progresso de 70%

# Aluno 2 participa de um treinamento de procedimento endoscópico
inicializar_treinamento(2, "procedimento endoscópico")
atualizar_progresso(2, 1, 80)  # Progresso de 80%

# Exibindo resultados
exibir_resultados()

# Visualização gráfica do progresso dos alunos (pode ser opcional na documentação)
import matplotlib.pyplot as plt

def visualizar_progresso():
    """
    Exibe um gráfico do progresso dos alunos em diferentes treinamentos.
    """
    alunos = np.arange(1, 11)  # IDs dos alunos (1 a 10)
    for i in range(progresso_treinamento.shape[1]):
        plt.plot(alunos, progresso_treinamento[:, i], label=f'Treinamento {i+1}')
    
    plt.xlabel('ID do Aluno')
    plt.ylabel('Progresso (%)')
    plt.title('Progresso dos Alunos nos Treinamentos')
    plt.legend()
    plt.show()

# Visualizar gráfico
visualizar_progresso()
```
